```python
df['TotalCharges'] = df['TotalCharges'].replace(" ", np.nan)
df['TotalCharges'] = df['TotalCharges'].astype(float)
```

```python
def convert(value):
    if value == 1:
        return 'Yes'
    else:
        return 'No'

df['SeniorCitizen'] = df['SeniorCitizen'].apply(convert)
```

```python
df.head()
```

```python
ax = sns.countplot(x = df['Churn'], data = df)
ax.bar_label(ax.containers[0])
plt.title('Count of Customer by Churn')
plt.show()
```

```python
gb = df.groupby('Churn').agg({'Churn':'count'})
plt.pie(gb['Churn'], labels = gb.index, autopct='%1.2f%%')
plt.title("Percentage of Churn")
plt.show()
```

```python
sns.countplot(x = df['gender'], data = df, hue = df['Churn'])

plt.title("Churn by Gender")
plt.show()
```

```python
# Calculate counts
ct = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Convert to percentage
pct = ct.div(ct.sum(axis=1), axis=0) * 100

# Plot stacked bar chart
ax = pct.plot(kind='bar', stacked=True)

# Add percentage labels
for c in ax.containers:
    ax.bar_label(c, fmt='%.1f%%', label_type='center')

plt.title("Churn by Senior Citizen")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage (%)")
plt.legend(title="Churn", bbox_to_anchor=(1.05, 1))
plt.tight_layout()
plt.show()
```

```python
 plt.figure(figsize=(10, 6))
sns.histplot(x = df['tenure'], data = df, bins = 50, hue = df['Churn'])
plt.title('Distribution of Tenure')
plt.show()
```

People who have used our services for one to two months have churned

```python
ax = sns.countplot(x = df['Contract'], data = df, hue = df['Churn'])
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.title("Churn by Contract")
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

plt.figure(figsize=(18, 20))   # adjust height if needed

for i, col in enumerate(cols, 1):
    plt.subplot(3, 3, i)       # 3 rows × 3 columns grid
    ax = sns.countplot(data=df, x=col, hue='Churn')

    # Add count labels on bars
    for container in ax.containers:
        ax.bar_label(container, fontsize=9)

    plt.title(f"Count Plot of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")

plt.tight_layout()
plt.show()
```

Customers who churn generally have lower adoption of add-on services such as OnlineSecurity, OnlineBackup, DeviceProtection, TechSupport, and Streaming services.
Fiber-optic Internet users show noticeably higher churn compared to DSL and no-internet-service groups.

```python
ax = sns.countplot(x = df['PaymentMethod'], data = df, hue = df['Churn'])
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.xticks(rotation = 45)
plt.title("Churn by Payment")
```
